```python
## Load all necessary packages

import warnings
warnings.filterwarnings("ignore")
warnings.filterwarnings("ignore", message=".*converge.*")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import itertools
import math
import os

















```

```python
## Load and prepare data
FILE = "Electric_Production.csv"
if not os.path.exists(FILE):
    raise FileNotFoundError(f"File not found: {FILE}")

df = pd.read_csv(FILE)
df['DATE'] = pd.to_datetime(df['DATE'], infer_datetime_format=True)
df = df.set_index('DATE').sort_index()
df = df.asfreq('MS')
value_col = df.columns[0]
series = df[value_col].astype(float)
```

```python

## Plot series
plt.figure(figsize=(12,4))
plt.plot(series, label=value_col)
plt.title("Electric Production (time series)")
plt.xlabel("Date")
plt.ylabel(value_col)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
## Decomposition
decomp = seasonal_decompose(series, model='additive', period=12, extrapolate_trend='freq')
plt.rcParams.update({'figure.figsize': (12,9)})
decomp.plot()
plt.suptitle("Seasonal Decomposition (Additive)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
## Moving Average
ma12 = series.rolling(window=12).mean()
plt.figure(figsize=(12,4))
plt.plot(series, label='Original')
plt.plot(ma12, label='12-month Moving Average', linewidth=2)
plt.title("Original Series and 12-Month Moving Average")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
## Exponential Smoothing
ses = SimpleExpSmoothing(series, initialization_method="estimated").fit()
ses_fitted = ses.fittedvalues
hw_add = ExponentialSmoothing(series, trend='add', seasonal='add', seasonal_periods=12, initialization_method="estimated").fit()
hw_add_fitted = hw_add.fittedvalues

plt.figure(figsize=(12,5))
plt.plot(series, label='Original')
plt.plot(ses_fitted, label='Simple Exp Smoothing')
plt.plot(hw_add_fitted, label='Holt-Winters Additive')
plt.title("Exponential Smoothing Fits")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
## Train/Test Split
test_periods = 36
train = series[:-test_periods]
test = series[-test_periods:]
```

```python
## SARIMA Grid Search 
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in pdq]

best_aic = np.inf
best_order = None
best_seasonal_order = None
best_model = None

print("Searching SARIMA hyperparameters ...")

for order in pdq:
    for seasonal_order in seasonal_pdq:
        try:
            mod = SARIMAX(train, order=order, seasonal_order=seasonal_order,
                          enforce_stationarity=False, enforce_invertibility=False)
            res = mod.fit(disp=False, maxiter=100)
            if res.aic < best_aic:
                best_aic = res.aic
                best_order = order
                best_seasonal_order = seasonal_order
                best_model = res
        except Exception:
            continue

print(f"Best SARIMA order: {best_order}, seasonal_order: {best_seasonal_order}, AIC: {best_aic:.2f}")
```

```python
## Fit and Forecast 

if best_model is None:
    print("Fallback to SARIMA(1,1,1)x(1,1,1,12)")
    res = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,12),
                  enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)
else:
    res = best_model

print(res.summary())

pred = res.get_forecast(steps=test_periods)
pred_mean = pred.predicted_mean
pred_ci = pred.conf_int()

rmse = math.sqrt(mean_squared_error(test, pred_mean))
print(f"\nTest RMSE: {rmse:.4f}")

plt.figure(figsize=(12,5))
plt.plot(train[-120:], label='Train (recent)')
plt.plot(test.index, test.values, label='Actual (test)', marker='o')
plt.plot(pred_mean.index, pred_mean.values, label='SARIMA Forecast', marker='o')
plt.fill_between(pred_ci.index, pred_ci.iloc[:,0], pred_ci.iloc[:,1], color='gray', alpha=0.3)
plt.title(f"SARIMA Forecast vs Actual (RMSE={rmse:.3f})")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
## Future Forecast
final_model = SARIMAX(series, order=res.model.order, seasonal_order=res.model.seasonal_order,
                      enforce_stationarity=False, enforce_invertibility=False)
final_res = final_model.fit(disp=False, maxiter=100)

future_steps = 36
future_pred = final_res.get_forecast(steps=future_steps)
future_mean = future_pred.predicted_mean
future_ci = future_pred.conf_int()

plt.figure(figsize=(12,5))
plt.plot(series[-240:], label='Recent history')
plt.plot(future_mean.index, future_mean.values, marker='o', label=f'Forecast ({future_steps} months)')
plt.fill_between(future_ci.index, future_ci.iloc[:,0], future_ci.iloc[:,1], alpha=0.25)
plt.title(f"Future Forecast — next {future_steps} months")
plt.legend()
plt.tight_layout()
plt.show()

print("\n--- SUMMARY ---")
print("Data span:", series.index.min().date(), "to", series.index.max().date())
print("Model:", res.model.order, "Seasonal:", res.model.seasonal_order)
print(f"Test RMSE: {rmse:.4f}")
```

```python
## Summary of findings

## . The forecast predicts fluctuations with clear seasonal patterns (rising during mid-year, falling at the start/end of the year).
## . The confidence intervals (CI) give the uncertainty range — 95% of the time, actual values should fall within those bands.
## .The overall trend continues upward, consistent with long-term growth in electric production.
```
